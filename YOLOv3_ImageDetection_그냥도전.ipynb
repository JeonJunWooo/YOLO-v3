## Detect Image 살펴보기

```python
!cd yolov3; python detect.py \
    --weights '/content/yolov3/pretrained/yolov3-fixed.pt' \
    --source '/content/yolov3/data/images' \
    --project '/content/yolov3/detected' \
    --name 'images' \
    --img 640 \
    --conf-thres 0.5 \
    --iou-thres 0.4 \
    --line-thickness 2 \
    --exist-ok \
    --device CPU
```
